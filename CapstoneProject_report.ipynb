We can use the k-means clustering algorithm for unsupervised machine learning like clustering.<br>
It covers practice of partitioning on data set into groups that have similar characteristics.<br>
K-means tries to minimize the intra-cluster distances; distance of data points from the centroid.<br>
K-means tries to maximize the inter-cluster distances; distance from other cluster centroid.<br>
K-means uses dissimilarity measures to find the k value meaning the number of clusters.<br>
The k value depend on the neighbourhood and the fact that we are trying to find the kind of<br> 
business to open. So for our case, we run k-means to cluster the neighbourhood into 5 clusters.<br>
